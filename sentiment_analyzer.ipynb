```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

def analyze_vader(text):
    sid = SentimentIntensityAnalyzer()

    score = sid.polarity_scores(text)['compound']

    if score > 0.5:
      res = "Very Positive! ğŸŒŸ"
    elif score > 0:
      res = "Slightly Positive ğŸ™‚"
    elif score == 0:
      res = "Neutral ğŸ˜"
    elif score > -0.5:
      res = "Slightly Negative ğŸ™"
    else:
      res = "Very Negative ğŸ’¢"

    return score, res


input_text = input("Please enter your text here: ")

s, r = analyze_vader(input_text)

print(f"Hi!!This is what you entered: {input_text}")
print(f"Here is the score: {s:.2f}")
print(f"Here is the result: {r}")
```
